```python
# Import Dependencies
import pandas as pd
```

```python
# Creating file path
census_file = "../Original_Data_Sets/Census_2020_income.csv"

# Reading resource csv file as DataFrame
census_file_df = pd.read_csv(census_file, encoding="UTF-8")

# Drop all rows with no values
census_file_df = census_file_df.dropna()

# Visualizing DataFrame
census_file_df.head()
```

```python
# Creating index_list to store column indicies we are keeping
# Keeping only county households income estimates
# Starting list at [0] to keep the Label(Grouping) column
index_list = [0]

# Starting calculation on column index 1: Washington State Households Estimate
n = 1

# 321 columns, starting count after column 1, deduct column 0 and 1
# Total columns to count = 319, 319/8 = 39.875, round down to 39
# For loop to calculate column indicies to keep, perform calculation 39 times
for x in range(39):
    # Keeping every 9th column after 1st column, hence n + 8
    n = n + 8

    # Append output to index_list
    index_list.append(n)
```

```python
# Creating new DataFrame from the column indicies saved in index_list
census_file_cleaned_df = census_file_df.iloc[:, index_list].copy()

# Visualizing new DataFrame
census_file_cleaned_df.head(20)
```

```python
# Setting index as "Label (Grouping)" column
census_income_draft_df = census_file_cleaned_df.set_index(census_file_cleaned_df["Label (Grouping)"])

# Dropping original "Label (Grouping)" column
census_income_draft_df.drop(columns=["Label (Grouping)"], inplace=True)

# Remove "Label (Grouping)" label from index column
census_income_draft_df = census_income_draft_df.rename_axis(None)

# Visualizing new DataFrame
census_income_draft_df.head()
```

```python
# Transpose DataFrame inverting columns and rows
census_income_df = census_income_draft_df.transpose()

# Reducing column to only "Median income (dollars)"
census_income_df = census_income_df[["Median income (dollars)"]]

# Reset index of DataFrame to duplicate index with old column labels
census_income_df = census_income_df.reset_index()

# Formating the old column labels, removing all string "," and after
census_income_df["index"] = census_income_df["index"].str.strip(",")

# Spliting county name from "County" in string
county_split = census_income_df["index"].str.split(" ", n=1, expand=True)

# Replacing old column labels with only the county name (no County after)
census_income_df["index"] = county_split[0]

# Rename the duplicate index column with cleaned up county names as "County"
census_income_df = census_income_df.rename(columns={"index":"County"})

# Setting "County" as the new index
census_income_df = census_income_df.set_index(census_income_df["County"])

# Dropping original "County column"
census_income_df.drop(columns=["County"], inplace=True)

# Visualize new DataFrame
census_income_df.head()
```

```python
# Remove "," from "Median income (dollars)"
census_income_df["Median income (dollars)"] = census_income_df["Median income (dollars)"].str.replace(",","")

# Convert "Median income (dollars)" to integers
census_income_df["Median income (dollars)"] = census_income_df["Median income (dollars)"].astype("int64")

# Visualize updated DataFrame
census_income_df.head()
```

```python
# Output cleaned DataFrame as csv file
census_income_df.to_csv("Cleaned_Data_Sets/Cleaned_Census_Income_Data.csv", index=True, header=True)
```
